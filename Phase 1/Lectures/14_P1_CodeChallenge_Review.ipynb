### 3b: What are some of the things we could do to deal with null values?
 

```python
# call_on_students(1)
```

Answer:

- Drop the nulls (2 ways)
    - either drop the column (if it has a lot of nulls)
    - drop the individual rows that contain nulls (if we won't lose too much data
- Fill in the nulls (2 ways)
    - Replace them with a reasonable value (a measure of central tendency)
    - Replace them with a common value for nulls like "Unknown"
- Create a missing indicator column - done in tandem with filling in nulls! 
    - a boolean new column on if data was null or not
   - obviously will be empty if you just dropped the nulls
